**Como resolver este problema?**

*   Os objetos (criérios/alternativas) são comparados paritáriamente em relação à outro objeto no nível imediatamente inferior da hierarquia
* As comparações são realizadas empregando a Escala Fundamental de Saaty

*   As comparações paritárias resultam em matrizes de comparações paritárias
*   Ao aplicar a Decomposição em Autovalores e Autovetores a essas matrizes, obtemos um autovetor, que, após normalização fornece os pesos (prioridades) dos objetos
* O produto vetorial entre os vetores de de prioridades locais, resulta no vetor de prioridade global que tem os pessos das Alternativas em relação ao Alvo





**Resolução Matemática**

Seja $A = (a_{ij}) _{N \times N}$ uma matriz de comparação paritária, obtidas após comparar objetos no nível $h$ com aqueles no nível $h - 1$.


Cada elemento $a_{ij}$ representa a preferência do objeto $i$ em relação ao objeto $j$, utilizando a escala numérica.

<b>Propriedade 1 (Identidade).</b> Isso acarreta que $a_{ii} = 1$.

<b>Propriedade 2 (Consistência).</b> Se a preferência entre $i$ e $j$ vale $2$, $a_{ij} = 2$, e a preferência entre $j$ e $k$ vale $3$, $a_{jk}=3$  , então espera-se que a preferência entre $i$ e $k$ valha $6$, $a_{ik} = 6$,  isto é, $a_{ik} = a_{ij} a_{jk}$.

<b>Propriedade 3 (Reciprocidade).</b> Como $a_{ik} = a_{ij} a_{jk}$, então $a_{ii} = a_{ij} a_{ji} ⇒ a_{ij} = 1/a_{ji}$.

Portanto, as matrizes de comparações paritárias são do tipo:
$$
A = \left[
\matrix{1 & a_{12} & a_{13} & ⋯ & a_{1N}\\ 1/a_{12} & 1 & a_{23} & ⋯ & a_{2N} \\ 1/a_{13} & 1/a_{23} & 1 & ⋯ & a_{3N}\\ \vdots & \vdots & \vdots & \ddots & \vdots\\ 1/a_{1N} & 1/a_{2N} & 1/a_{3N} & ⋯ & 1}
  \right]
$$

De acordo com Saaty as preferências absolutas $a_{ij}$ são as razões entre as preferências relativas $w_i$ e $w_j$, isto é, $a_{ij} = w_i / w_j$.

O objetito é então obter as preferências relativas, que formam o vetor
$$
w = [w_1, w_2, w_3, ⋯, w_N]^{T}
$$
Para tanto escreve-se a matriz $A$ da seguinte forma:
$$
A = \left[
\matrix{w_1/w_1 & w_1/w_2 & w_1/w_3 & ⋯ & w_1/w_N\\ w_2/w_1 & w_2/w_2 & w_2/w_3 & ⋯ & w_2/w_N \\ w_3/w_1 & w_3/w_2 & w_3/w_3 & ⋯ & w_3/w_N\\ \vdots & \vdots & \vdots & \ddots & \vdots\\ w_N/w_1 & w_N/w_2 & w_N/w_3 & ⋯ & w_N/w_N}
  \right]
$$

Agora, calculamos o produto $A w$, ou seja,
$$
Aw = \left[
\matrix{w_1/w_1 & w_1/w_2 & w_1/w_3 & ⋯ & w_1/w_N\\ w_2/w_1 & w_2/w_2 & w_2/w_3 & ⋯ & w_2/w_N \\ w_3/w_1 & w_3/w_2 & w_3/w_3 & ⋯ & w_3/w_N\\ \vdots & \vdots & \vdots & \ddots & \vdots\\ w_N/w_1 & w_N/w_2 & w_N/w_3 & ⋯ & w_N/w_N}
  \right] \left[\matrix{w_1 \\ w_2 \\ w_3 \\ \vdots \\ w_N} \right] = \left[\matrix{w_1 + w_1 + w_1 + ⋯ + w_1 \\ w_2 + w_2 + w_2 + ⋯ + w_2 \\ w_3 + w_3 + w_3 + ⋯ + w_3 \\ \vdots \\ w_N + w_N + w_N + ⋯ + w_N} \right] ⇒ Aw = nw
$$

Portanto, $w$ é um autovetor da matriz $A$ associado ao autovalor $n$. Assim, a solução é dada em termos da Decoposição em Autovalores e Autovetores da matriz $A$.

Após a obtenção do autovetor $w$ (também denominado de vetor de prioridade local ou global) este deve ser normalizado de tal modo que
$$
\sum_{i=1}^{N} w_i = 1
$$

O cálculo das matrizes de comparações paritárias deve ser realizado comparando todos os objetos em um nível da hierárquia com aqueles no nível inferior.

Assim, se temos $P$ alternativas e $K$ critérios, então teremos $K$ vetores de prioridades locais com $P$ elementos, resultantes das comparações das alternativas em relação aos critério. Estes vetores formarão um nova matriz $M_{K \times P}$ onde cada linha será formada pelos vetores locais relativos aos critérios. E também teremos um vetor $v_{1 \times K}$ de prioridade local resultante da comparação dos critérios em relação ao Alvo.

Por fim, para obtermos o vetor de prioridade global $u$, que fornece as preferências das Alternativas em relação ao Alvo, fazemos:
$$
u = vM
$$


**Verificando a consistência**

Para determinar a consistência da matriz de comparação paritária calculamos o índice de consistência
$$
CI = \dfrac{\lambda_{max} - N}{N-1}
$$

Se a razão $CI/RI < 10\%$ dizemos que a matriz é consistente, onde $RI$ é dado de acordo com a Tabela:

```python
import numpy as np
```

```python
'''
Comparações paritárias dos critérios
'''
A = np.eye(4)

#Retorno Mensal: 1
#IR: 2
#Perído: 3
#Tx. Adm.: 4

#Retorno Mensal x IR
A[0,1] = 2                   #a_12
A[1,0] = 1/A[0,1]            #a_21 = 1/a_12

#Retorno Mensal x Perído
A[0,2] = 3
A[2,0] = 1/A[0,2]

#Retorno Mensal x Tx. Adm.
A[0,3] = 5
A[3,0] = 1/A[0,3]

#IR x Período
A[1,2] = 4
A[2,1] = 1/A[1,2]

#IR x Tx. Adm.
A[1,3] = 5
A[3,1] = 1/A[1,3]

#Perído x Tx. Adm.
A[2,3] = 3
A[3,2] = 1/A[2,3]

print(A)
```

```python
'''
Função que retorna se a matriz é ou não cosnsistente
'''
def CIRI(lamb, N):
    ri = [0, 0, 0.52, 0.89, 1.11, 1.25, 1.35, 1.40, 1.45, 1.49, 1.52, 1.54, 1.56, 1.58, 1.59]
    ci = (lamb - N) / (N - 1)
    cr = ci / ri[N]
    if cr > 0.1:
        print('Inconsistente: %.2f' %cr)
    else:
        print('Consistente: %.2f' %cr)

'''
Função que calcula o autovetor retornando-o juntamente com o autovalor
'''
def VV(A):
    l, v = np.linalg.eig(A)
    v = v.T
    i = np.where(l == np.max(l))[0][0]
    l = l[i]
    v = v[i]
    v = v/np.sum(v)
    return np.real(l), np.real(v)
```

```python
'''
Obtém o autovetor e autovalor
Calcula a consistência
'''
l, v = VV(A)

print('Autovalor: %.2f' %l)
print('Autovetor: ', np.round(v, 2))

CIRI(l, A.shape[0])
```

**Comparações paritárias das Alternativas**

Agora é necessário avaliar cada uma das grandezas, se quanto maior seu valor mais preferível ou menos preferível ela é. Isso é necessário, porque dependendo da resposta uma fórmula de normalização será utilizada.

*   Retorno mensal (%): quanto maior, mais preferível
*   IR: quanto maior, mais preferível
*   Período (anos): quanto menor, mais preferível
*   Tx. Adm (% a.a): quanto menor, mais preferível

Para grandezas do tipo <b>quanto maior, mais preferível</b> a normalização é feita assim:
$$
{v'}_{ij}=\frac{v_{ij}}{\sum_{k=1}^{K}v_{kj}}
$$

Para grandezas do tipo <b>quanto menor, mais preferível</b> a normalização é feita assim:
$$
{v'}_{ij}=\frac{1}{v_{ij}\sum_{k=1}^{K}\frac{1}{v_{kj}}}
$$


```python
'''
Normalização: grandezas do tipo quanto maior, mais preferível
'''
def NormMaior(v):
    v = np.array(v)
    return v/np.sum(v)

'''
Normalização: grandezas do tipo quanto menor, mais preferível
'''
def NormMenor(v):
    v = np.array(v)
    return 1 / (v * np.sum(1/v))
```

```python
'''
Dados dos investimentos
'''
Retorno_mensal = [0.7, 0.9, 0.6, 0.7]
IR             = [1, 1, 2, 2]
Periodo        = [3, 5, 3, 4]
Tx_Adm         = [0.1, 2.1, 1.1, 0.1]

Retorno_mensal_norm = NormMaior(Retorno_mensal)
IR_norm             = NormMaior(IR)
Periodo_norm        = NormMenor(Periodo)
Tx_Adm_norm         = NormMenor(Tx_Adm)

M = np.array([Retorno_mensal_norm, IR_norm, Periodo_norm, Tx_Adm_norm])

print('Matriz normalizada')
print(M.T)
```

```python
'''
Obtém o vetor de piroridade global
'''
u = np.dot(v, M.T)

print('Prioridades das Opções de Investimentos:')
print('Opção A: %.4f' %u[0])
print('Opção B: %.4f' %u[1])
print('Opção C: %.4f' %u[2])
print('Opção D: %.4f' %u[3])

print('\nAlocação do total disponível de R$ 100.000,00:')
total = 100000
print('Opção A: %.2f' %(u[0] * total))
print('Opção B: %.2f' %(u[1] * total))
print('Opção C: %.2f' %(u[2] * total))
print('Opção D: %.2f' %(u[3] * total))
```
